```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
d = {'Actual values': [87,17],'Predicted':[16, 59]}
df = pd.DataFrame(data=d)
df
```

```python
print(df.info)
print(" ")
print(df.info())
```

```python
score = (87 + 59)/ (87 + 59 + 16 +17)
score
```

```python
sns.heatmap(df, cmap = 'Blues', annot= True, xticklabels= '01', yticklabels= '01')
plt.xlabel('Predicted label')
plt.ylabel('Actual label')
plt.title('Accuracy Score:{}'.format(score))
```
